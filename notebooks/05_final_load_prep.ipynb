```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/final_cleaned_spotify_data (1).csv')
print(f'Final Dataset: {df.shape[0]:,} rows x {df.shape[1]} columns')
```

```python
print('=== Null Values ===')
print(f"Total nulls: {df.isnull().sum().sum()}")

print('\n=== Target Variable Distribution (Final Status) ===')
# Iterating through the Final_status categories
for status in df['Final_status'].unique():
    count = (df['Final_status'] == status).sum()
    pct = (df['Final_status'] == status).mean() * 100
    print(f"{status}: {count:,} ({pct:.2f}%)")

print('\n=== Column Data Types ===')
print(df.dtypes.value_counts())

print('\n=== KPI Summary ===')
print(f"Overall Churn Rate: {df['is_churned'].mean()*100:.2f}%")
print(f"Average Listening Hours/Week: {df['avg_listening_hours_per_week'].mean():.2f} hours")
print(f"Average Tenure: {df['tenure_months'].mean():.2f} months")
print(f"Average Engagement Score: {df['engagement_score'].mean():.4f}")
```

```python
print('=== Column Information ===')
for i, col in enumerate(df.columns, 1):
    print(f'{i:2d}. {col:40s} | {str(df[col].dtype):8s} | {df[col].nunique():6,} unique values')
```
